## Visualization 1: Average Birth Rate and Death Rate for countries across all the years
- A bar plot is a good option for visualizing average birth and death rates country-wise because it allows for easy comparison of the rates between different countries. The length of each bar represents the value of the rate, making it easy to see which countries have higher or lower rates.

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'yellow', 'brown']

def group_df(feature):
    
    """
    Groups the dataset by country and calculates the average value of the given feature.
    Feature can be Birth Rate and Death Rate. It will return a pandas.DataFrame, 
    a new dataframe with the average value of the feature for each country.
    """
    
    # create a new dataframe
    df_grouped=pd.DataFrame()

    # find average for each country
    df_grouped['Avg. ' + feature]=dataset.groupby('Country')[feature].mean()

    # set the index as a column - countries
    df_grouped['Country']=df_grouped.index

    # drop the index
    df_grouped.reset_index(drop=True, inplace=True)

    # sort the rows based of Avg Birth rate
    df_grouped.sort_values('Avg. '+feature, inplace=True, ascending=False)

    print("Avg. " + feature)
    display(df_grouped)
    
    return df_grouped

def plot_bar_graph(dataframe, x_feature, y_feature):
    
    """
    Creates a bar graph using the given dataframe and x and y feature values.
    
    Args:
    - dataframe: pandas.DataFrame, the dataframe to use for the plot
    - x_feature: str, the name of the feature to use for the x-axis
    - y_feature: str, the name of the feature to use for the y-axis
    """
    
    # Group the dataframe by the x feature and calculate the mean of the y feature for each group
    grouped_data = dataframe.groupby(x_feature)[y_feature].mean()
    
    # Create a bar chart using the grouped data
    plt.bar(grouped_data.index, grouped_data.values, color = colors)
    
    # Set the labels for the x and y axes
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    
    # Show the plot
    plt.show()
```

```python
df_birth=group_df('Birth Rate')
plot_bar_graph(df_birth, 'Country', 'Avg. Birth Rate')

print("========================================================")

df_death=group_df('Death Rate')
plot_bar_graph(df_death, 'Country', 'Avg. Death Rate')
```

Based on the two plots presented above, it is apparent that India has the highest average birth and death rates, despite being the second most populous country in 2018. It is noteworthy that South Africa exhibits a similar trend, which is surprising given its relatively small population. In contrast, China has a higher birth rate than death rate, contributing to its position as the most populous country in the world. Conversely, the United Kingdom has a significantly higher death rate compared to the average birth rate, leading to a relatively small population.

## Visualization 2: GDP for all countries in the last 10 years

- A multi-line plot is best for this situation because it allows us to compare the trends in GDP over time for multiple countries on the same plot, making it easy to identify similarities and differences between countries. It also allows us to easily see how the GDP of each country has changed over time relative to the other countries, providing a more comprehensive view of the data than a single line plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

def gdp_multi_line_plot(df):
    """
    Create a multi-line plot of GDP over time for different countries,
    using data from a dataframe. 
    
    Args:
    - dataset

    """

    # Create figure and axis objects
    fig, ax = plt.subplots(figsize=(15, 10))

    # Get unique values for the Country feature
    countries = df['Country'].unique()

    # Plot each line for a different country
    for country in countries:
        df_subset = df[df['Country'] == country]
        ax.plot(df_subset['Year'], df_subset['GDP in USD'], label=country)

    # Set the chart title and axis labels
    ax.set_title('GDP Over Time by Country', fontsize=18)
    ax.set_xlabel('Year', fontsize=14)
    ax.set_ylabel('GDP in USD', fontsize=14)

    # Add a legend and show the plot
    ax.legend()
    plt.show()

gdp_multi_line_plot(dataset)
```

- A line plot clearly depicts trends over time and can also be used to compare the patterns of several categories. We can see that the United States has the greatest GDP of all countries, which has been consistent over time. 
- China's GDP was low in 2008 and has increased significantly since then, owing to their growth in numerous industries such as manufacturing; but, in comparison to the United States, it is still quite low.
- The GDP of Japan shows a modest increase up until 2012, which could be attributable to Japan's growth plan, which was implemented to lift the country out of deflation, but then it drops because of earthquake i guess.


## Visualization 3: Electric Power Consumption vs Population for India and China
- A scatter plot is the best choice for this scenario because it allows us to visualize the relationship between two continuous variables (Total Population and Electric Power Consumption) for multiple categories (Country) at once, making it easy to identify any patterns or trends. Additionally, the use of markers for each country allows for easy differentiation between the data points, while the legend provides additional information on each country's data.


```python
import pandas as pd
import matplotlib.pyplot as plt

def power_consumption_scatter_plot(df):
    """
    Create a scatter plot of electric power consumption per capita versus
    total population for different countries, using data from a CSV file.

    Args:
    - dataset

    """

    # Create figure and axis objects
    fig, ax = plt.subplots(figsize=(15, 10))

    # Get unique values for the Country feature
    countries = df['Country'].unique()

    # Create a scatter plot for each country
    for country in countries:
        df_subset = df[df['Country'] == country]
        ax.scatter(df_subset['Total Population'], df_subset['Electric Power Consumption(kWH per capita)'], label=country)

    # Set the chart title and axis labels
    ax.set_title('Electric Power Consumption vs Total Population by Country', fontsize=18)
    ax.set_xlabel('Total Population', fontsize=14)
    ax.set_ylabel('Electric Power Consumption (kWH per capita)', fontsize=14)

    # Add a legend and show the plot
    ax.legend()
    plt.show()

power_consumption_scatter_plot(dataset)
```

- The graph that accompanies this text makes it clear that electricity demand will increase alongside population growth. 
- China's considerable surge in electric power usage serves as evidence of this correlation.
- Meanwhile, India has experienced a gradual increase in consumption in tandem with population growth.
- The constant numbers on both lines signify the mean consumption.

## Visualization 4: Employement Percentage in the Agriculture
- Pie plots are a good option for showing proportions or percentages of different categories, making them suitable for displaying the relative sizes of employment in agriculture across different countries. They are also visually appealing and can easily convey the top 10 countries with the highest employment in agriculture for each year.

```python
import csv
import matplotlib.pyplot as plt

def plot_agriculture_employment(csv_file):
    """
    Plots the employment in agriculture percentage for the top 10 countries for the years 2015
    to 2018. It takes a CSV file path as input, reads the data for the years 2015 to 2018,
    extracts the employment in agriculture percentage for each country in each year, and
    then plots a pie chart for the top 10 countries with the highest employment in agriculture
    for each year in a 2x2 grid. The function does not return any value.

    Args:
        csv_file (str): The path to the CSV file containing the data.

    Returns:
        None.
    """
    
    # Load the data from the CSV file
    data_2015 = {}
    data_2016 = {}
    data_2017 = {}
    data_2018 = {}
    with open(csv_file, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            if row['Year'] == '2015':
                data_2015[row['Country']] = float(row['Employment in Agriculture(%)'])
            elif row['Year'] == '2016':
                data_2016[row['Country']] = float(row['Employment in Agriculture(%)'])
            elif row['Year'] == '2017':
                data_2017[row['Country']] = float(row['Employment in Agriculture(%)'])
            elif row['Year'] == '2018':
                data_2018[row['Country']] = float(row['Employment in Agriculture(%)'])

    # Plot the data in four subplots for each year
    fig, axs = plt.subplots(2, 2, figsize=(15, 15))
    axs = axs.flatten()

    for i, (year, data) in enumerate(zip(['2015', '2016', '2017', '2018'], [data_2015, data_2016, data_2017, data_2018])):
        # Get the top 10 countries with the highest employment in agriculture
        sorted_data = sorted(data.items(), key=lambda x: x[1], reverse=True)[:10]
        countries = [x[0] for x in sorted_data]
        values = [x[1] for x in sorted_data]

        # Plot the pie chart
        axs[i].pie(values, labels=countries, autopct='%1.1f%%')
        axs[i].set_title(f'Employment in Agriculture (%) in {year}')
        axs[i].legend(loc='upper right', bbox_to_anchor=(1.2, 1))
        axs[i].axis('equal')  # Make the pie chart circular
        axs[i].grid()  # Add a grid

    # Show the plot
    plt.show()

plot_agriculture_employment('dataset/worldbankindicatorsdatase.csv')
```

- From the graph above, India has the biggest percentage of people employed in agriculture out of all countries. This reflects the fact that India is a predominantly agricultural country with numerous food products grown on its soil. India is followed by China. 
- Due to a scarcity of arable land in China, farming is particularly labor intensive. As a result, agriculture employs many people. Agriculture employs a very small percentage of the population in the remainder of the world. The United Kingdom has the lowest percentage in agriculture, which is since it produces less than 60% of the food consumed.
